# PokeGAN Dataset Preprocessing

```python
import os
import pandas as pd
from PIL import Image
```

```python
def get_type_of_pokemon(pokemon_name, pokemon_list):
    for index, row in pokemon_list.iterrows():
        if row["Name"] == pokemon_name:
            return row["Type 1"]
```

```python
## Directory For Loop

directory = os.getcwd()
pokemon_data_directory = directory + "/PokemonData"
pokemon_list = pd.read_csv("Pokemon.csv")
```

```python
name_list = pokemon_list["Name"].tolist()
for folder in os.listdir(pokemon_data_directory):
    if folder in name_list:
        pokemon_type = get_type_of_pokemon(folder, pokemon_list)
        folder_directory = pokemon_data_directory + "/" + folder
        for image_filename in os.listdir(folder_directory):
            if ".jpg" in image_filename or ".jpeg" in image_filename or ".png" in image_filename:
                print(directory, pokemon_type, image_filename)
                if pokemon_type is not None:
                    directory_to_open_image = directory + "/PokemonData/" + folder + "/" + image_filename
                    directory_to_be_saved = directory + "/PokemonTypeImages/" + pokemon_type + "/" + image_filename + "_256.png"
                    image = Image.open(directory_to_open_image)
                    image_resized = image.resize((256, 256))
                    image_resized.save(directory_to_be_saved)
```
